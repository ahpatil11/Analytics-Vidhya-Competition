```python
train['Salary_Change'] = train.groupby('Emp_ID')['Salary'].diff().gt(0)
```

```python
train['MMM-YY'] = pd.to_datetime(train['MMM-YY'])
train['LastWorkingDate'] = pd.to_datetime(train['LastWorkingDate'])
train['Dateofjoining'] = pd.to_datetime(train['Dateofjoining'])
```

```python
(train['MMM-YY'] - train['Dateofjoining']) // np.timedelta64(1, 'M')
```

```python
train['Employed_Time'] = (train['MMM-YY'] - train['Dateofjoining']) // np.timedelta64(1, 'M')
```

```python
np.where(train['Employed_Time'] < 0, 0, train['Employed_Time'])
```

```python
train['Employed_Time'] = np.where(train['Employed_Time'] < 0, 0, train['Employed_Time'])
```

```python
train['Designation_change'] = train['Designation'] - train['Joining Designation']
```

```python
employed_time = train.groupby(by = 'Emp_ID', axis = 0)['Employed_Time'].max().reset_index()
employed_time.head()
```

```python
rating_avg = train.groupby(by = 'Emp_ID', axis = 0).mean()['Quarterly Rating']
rating_avg = rating_avg.reset_index()
rating_avg.head()
```

```python
business_value = train.groupby(by = 'Emp_ID', axis = 0).mean()['Total Business Value'].reset_index()
business_value.head()
```

```python
age = train.groupby(by = 'Emp_ID', axis = 0)['Age'].max().reset_index()
age.head()
```

```python
salary = train.groupby(by = 'Emp_ID', axis = 0)['Salary'].max().reset_index()
salary.head()
```

```python
gender = train.groupby(by = 'Emp_ID', axis = 0)['Gender'].agg(lambda x:x.value_counts().index[0]).reset_index()
gender.head()
```

```python
education = train.groupby(by = 'Emp_ID', axis = 0)['Education_Level'].agg(lambda x:x.value_counts().index[0]).reset_index()
education.head()
```

```python
joining_desg = train.groupby(by = 'Emp_ID', axis = 0)['Joining Designation'].min().reset_index()
joining_desg.head()
```

```python
designation = train.groupby(by = 'Emp_ID', axis = 0)['Designation'].min().reset_index()
designation.head()
```

```python
from functools import reduce
```

```python
df = reduce(lambda x,y: pd.merge(x,y, on='Emp_ID', how='inner'), [rating_avg, business_value, age, salary, gender, education,
                                                                 joining_desg, designation, employed_time])
```

```python
df['Designation_change'] = df['Designation']-df['Joining Designation']
```
