```python
train['Health Score'] = train['Health Score'].fillna(0)
test['Health Score'] = test['Health Score'].fillna(0)
```

```python
train['Health_Score'] = train['Health_Score'].fillna(0)
test['Health_Score'] = test['Health_Score'].fillna(0)
```

```python
train['Number_of_stall_visited'] = train['Number_of_stall_visited'].fillna(0)
test['Number_of_stall_visited'] = test['Number_of_stall_visited'].fillna(0)
```

```python
train['Last_Stall_Visited_Number'] = train['Last_Stall_Visited_Number'].fillna(0)
test['Last_Stall_Visited_Number'] = test['Last_Stall_Visited_Number'].fillna(0)
```

```python
train['Result'] = (train['Health_Score'] + train['Health Score']) + (train['Number_of_stall_visited'] + train['Last_Stall_Visited_Number']/30)
```

```python
train['City_Type'] = train['City_Type'].fillna(train['City_Type'].value_counts().index[0])
test['City_Type'] = test['City_Type'].fillna(test['City_Type'].value_counts().index[0])
```

```python
train['Employer_Category'] = train['Employer_Category'].fillna(train['Employer_Category'].value_counts().index[0])
test['Employer_Category'] = test['Employer_Category'].fillna(test['Employer_Category'].value_counts().index[0])
```

```python
train = train.drop(['Unnamed: 4', 'Donation','Health_Score', 'Health Score', 'Number_of_stall_visited', 'Last_Stall_Visited_Number'], axis = 1)
test = test.drop(['Unnamed: 4', 'Donation','Health_Score', 'Health Score', 'Number_of_stall_visited', 'Last_Stall_Visited_Number'], axis = 1)
```

```python
train = train.drop(['Registration_Date', 'Camp_Start_Date','Camp_End_Date','First_Interaction'], axis = 1)
test = test.drop(['Registration_Date', 'Camp_Start_Date','Camp_End_Date','First_Interaction'], axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['City_Type'] = le.fit_transform(train['City_Type'])
train['Employer_Category'] = le.fit_transform(train['Employer_Category'])
train['Category1'] = le.fit_transform(train['Category1'])
train['Category2'] = le.fit_transform(train['Category2'])
train['Category3'] = le.fit_transform(train['Category3'])
train['Income'] = le.fit_transform(train['Income'])
train['Education_Score'] = le.fit_transform(train['Education_Score'])
train['Age'] = le.fit_transform(train['Age'])

test['City_Type'] = le.fit_transform(test['City_Type'])
test['Employer_Category'] = le.fit_transform(test['Employer_Category'])
test['Category1'] = le.fit_transform(test['Category1'])
test['Category2'] = le.fit_transform(test['Category2'])
test['Category3'] = le.fit_transform(test['Category3'])
test['Income'] = le.fit_transform(test['Income'])
test['Education_Score'] = le.fit_transform(test['Education_Score'])
test['Age'] = le.fit_transform(test['Age'])
```

```python
x_train = train.drop('Result', axis = 1)
y_train = train['Result']
x_test = test
```

```python
from lightgbm import LGBMRegressor
lgb = LGBMRegressor()
y_pred = lgb.fit(x_train, y_train).predict(x_test)
solution = pd.DataFrame({'Patient_ID': test.Patient_ID, 'Health_Camp_ID': test.Health_Camp_ID, 'Outcome': y_pred})
solution.to_csv('Solution.csv', index = False)
```
