```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
train['Hospital_type_code'] = le.fit_transform(train['Hospital_type_code'])
train['Hospital_region_code'] = le.fit_transform(train['Hospital_region_code'])
train['Department'] = le.fit_transform(train['Department'])
train['Ward_Type'] = le.fit_transform(train['Ward_Type'])
train['Ward_Facility_Code'] = le.fit_transform(train['Ward_Facility_Code'])
train['Type of Admission'] = le.fit_transform(train['Type of Admission'])
train['Severity of Illness'] = le.fit_transform(train['Severity of Illness'])

test['Hospital_type_code'] = le.fit_transform(test['Hospital_type_code'])
test['Hospital_region_code'] = le.fit_transform(test['Hospital_region_code'])
test['Department'] = le.fit_transform(test['Department'])
test['Ward_Type'] = le.fit_transform(test['Ward_Type'])
test['Ward_Facility_Code'] = le.fit_transform(test['Ward_Facility_Code'])
test['Type of Admission'] = le.fit_transform(test['Type of Admission'])
test['Severity of Illness'] = le.fit_transform(test['Severity of Illness'])
```

```python
train.duplicated().sum()
```

```python
train['Bed Grade'] = train['Bed Grade'].fillna(train['Bed Grade'].mean())
test['Bed Grade'] = test['Bed Grade'].fillna(test['Bed Grade'].mean())
```

```python
train['City_Code_Patient'] = train['City_Code_Patient'].fillna(train['City_Code_Patient'].mean())
test['City_Code_Patient'] = test['City_Code_Patient'].fillna(test['City_Code_Patient'].mean())
```

```python
def age_fe(x):
    if x == '0-10':
        return 0
    elif x == '11-20':
        return 1
    elif x == '21-30':
        return 2
    elif x == '31-40':
        return 3
    elif x == '41-50':
        return 4
    elif x == '51-60':
        return 5
    elif x == '61-70':
        return 6
    elif x == '71-80':
        return 7
    elif x == '81-90':
        return 8
    else:
        return 9
```

```python
train['Age'] = train['Age'].apply(age_fe)
test['Age'] = test['Age'].apply(age_fe)
```

```python
train['Stay'].value_counts()
```

```python
def stay_fe(x):
    if x == '0-10':
        return 0
    elif x == '11-20':
        return 1
    elif x == '21-30':
        return 2
    elif x == '31-40':
        return 3
    elif x == '41-50':
        return 4
    elif x == '51-60':
        return 5
    elif x == '61-70':
        return 6
    elif x == '71-80':
        return 7
    elif x == '81-90':
        return 8
    elif x == '91-100':
        return 9
    else:
        return 10
```

```python
train['Stay'] = train['Stay'].apply(stay_fe)
```

```python
newtrain = train.copy()
newtest = test.copy()
```

```python
x_train = train.drop('Stay', axis = 1)
y_train = train['Stay']
x_test = test
```

```python
from sklearn.feature_selection import SelectKBest,chi2
bestFeatures = SelectKBest(score_func=chi2, k = 17)
fit = bestFeatures.fit(x_train, y_train)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x_train.columns)

featureScores = pd.concat([dfcolumns, dfscores], axis = 1)
featureScores.columns = ['Specs','Score']

featureScores.nlargest(14, 'Score')
```
